```python
from textblob import TextBlob
import pandas as pd

# Sample DataFrame creation for demonstration purposes
# In a real scenario, you would load your data into 'df'
data = {'review': ['This is a great product!', 'I am not happy with this.', 'It works as expected.']}
df = pd.DataFrame(data)

def get_sentiment(text):
    return TextBlob(str(text)).sentiment.polarity

df['sentiment_score'] = df['review'].apply(get_sentiment)

df['sentiment'] = df['sentiment_score'].apply(
    lambda x: 'positive' if x > 0 else 'negative' if x < 0 else 'neutral'
)

df[['review','sentiment','sentiment_score']].head()
```

```python
df['sentiment'].value_counts().plot(kind='bar')
```

```python
import seaborn as sns
sns.countplot(x='sentiment', data=df)
```
